```python
import pandas as pd
import numpy as np
# --- pandas_flavor compatibility shim (run this before importing janitor) ---
import pandas_flavor as pf
# If groupby decorator doesn't exist, alias it to the dataframe decorator
if not hasattr(pf, "register_groupby_method") and hasattr(pf, "register_dataframe_method"):
    pf.register_groupby_method = pf.register_dataframe_method
import janitor
from janitor import clean_names
from pathlib import Path
```

```python
try:
    from janitor import clean_names as jn_clean_names
    HAS_JANITOR = True
except Exception as e:
    HAS_JANITOR = False
    print(f"pyjanitor unavailable or failed to import: {e}\nUsing local clean_names fallback.")

def clean_names_fallback(df: pd.DataFrame) -> pd.DataFrame:
    def _clean(s):
        s = str(s).strip().lower().replace(" ", "_").replace("-", "_")
        s = s.encode("ascii", "ignore").decode()
        while "__" in s:
            s = s.replace("__", "_")
        return s
    return df.rename(columns=_clean)
```

```python
# little function to define the file root on different machines
def find_qm_root(start_path: Path = Path.cwd(), anchor: str = "QM") -> Path:
    """
    Traverse up from the start_path until the anchor folder (e.g., 'QM_Fork')      is found. Returns the path to the anchor folder.
    """
    for parent in [start_path] + list(start_path.parents):
        if parent.name == anchor:
            return parent
    raise FileNotFoundError(f"Anchor folder '{anchor}' not found in path      hierarchy.")
  
qm_root = find_qm_root()
```

```python
# Absence data
england_abs = pd.read_csv(base_path / "2022-2023" / "england_abs.csv", na_values=na_all, dtype={"URN": str})
```

```python
# Census data
england_census = pd.read_csv(base_path / "2022-2023" / "england_census.csv", na_values=na_all, dtype={"URN": str})
england_census.iloc[:, 4:23] = england_census.iloc[:, 4:23].apply(lambda col: pd.to_numeric(col.astype(str).str.replace('%', '', regex=False), errors="coerce"))
```

```python
# KS4 pupil destination data
england_ks4_pupdest = pd.read_csv(base_path / "2022-2023" / "england_ks4-pupdest.csv", na_values=na_all, dtype={"URN": str})
england_ks4_pupdest.iloc[:, 7:82] = england_ks4_pupdest.iloc[:, 7:82].apply(lambda col: pd.to_numeric(col.astype(str).str.replace('%', '', regex=False), errors="coerce"))
```

```python
# KS4 final attainment data
england_ks4final = pd.read_csv(base_path / "2022-2023" / "england_ks4final.csv", na_values=na_all, dtype={"URN": str})
start_col = "TOTPUPS"
end_col = "PTOTENT_E_COVID_IMPACTED_PTQ_EE"
cols_range = england_ks4final.loc[:, start_col:end_col].columns
# Strip % signs and convert to numeric
england_ks4final[cols_range] = england_ks4final[cols_range].apply(
    lambda col: pd.to_numeric(col.astype(str).str.replace('%', '', regex=False), errors="coerce")
)
```

```python

# School information data
england_school_information = pd.read_csv(
    base_path / "2022-2023" / "england_school_information.csv",
    na_values=na_all,
    dtype={"URN": str},
    parse_dates=["OFSTEDLASTINSP"],
    dayfirst=True  # Adjust if needed
)
```

```python
# 把多张表按学校唯一编号拼成一张“英国学校主表”（master file），用的都是 left join
# Left join england_ks4final with england_abs, census, and school information
england_school_2022_23 = (
    england_ks4final
    .merge(england_abs, on="URN", how="left")
    .merge(england_census, on="URN", how="left")
    .merge(england_school_information, on="URN", how="left")
    .merge(edubase_schools, left_on="URN", right_on="urn", how="left")
)
```

```python
# 导入画图底层库 matplotlib。seaborn（更方便画统计图，基于 matplotlib）。numpy（生成序列、算正态曲线）。
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Compute statistics。中位数、均值、标准差。
median_value = england_school_2022_23["ATT8SCR"].median(skipna=True)
mean_value = england_school_2022_23["ATT8SCR"].mean(skipna=True)
sd_value = england_school_2022_23["ATT8SCR"].std(skipna=True)

# 新建画布，大小 10×6 英寸。
plt.figure(figsize=(10, 6))

# Histogram with density
# 用 seaborn 画 直方图：
# data=... 指定数据表
# x="ATT8SCR" 指定画哪一列
# stat="density"：y 轴不是“人数”，而是“密度”（总面积=1），方便和曲线叠加
# binwidth=1：每个柱子的宽度是 1 分
# color/alpha/linewidth/edgecolor：控制颜色、透明度、边框
sns.histplot(
    data=england_school_2022_23,
    x="ATT8SCR",
    stat="density",
    binwidth=1,
    color="#4E3C56",
    alpha=0.4,
    linewidth=0.5,
    edgecolor="white"
)

# Overlay normal distribution curve。在 ATT8SCR 的最小值到最大值之间生成 500 个均匀点，用来画平滑曲线。
x_vals = np.linspace(
    england_school_2022_23["ATT8SCR"].min(),
    england_school_2022_23["ATT8SCR"].max(),
    500
)
# 用均值 mean_value 和标准差 sd_value 计算一条正态分布密度曲线（这是公式直接算出来的），用来对比你的数据分布是否接近正态。
normal_density = (
    1 / (sd_value * np.sqrt(2 * np.pi))
) * np.exp(-0.5 * ((x_vals - mean_value) / sd_value) ** 2)
# 把这条正态曲线画在直方图上。
plt.plot(x_vals, normal_density, color="#2E6260", linewidth=1)

# Add vertical lines for median and mean
plt.axvline(median_value, color="black", linestyle="dotted", linewidth=1)
plt.axvline(mean_value, color="#F9DD73", linestyle="solid", linewidth=1)

# Annotate median and mean
plt.text(median_value, plt.gca().get_ylim()[1]*0.95, f"Median = {median_value:.1f}",
         color="black", ha="center", va="top", fontsize=10)
plt.text(mean_value, plt.gca().get_ylim()[1]*0.85, f"Mean = {mean_value:.1f}",
         color="#F9DD73", ha="center", va="top", fontsize=10)

# Customize labels and theme
plt.title("Attainment 8 - All Schools England and Wales, 2022/23 Academic Year")
plt.xlabel("Attainment 8 Score")
plt.ylabel("Density")
sns.despine()
plt.tight_layout()

# Show the plot
# plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(12, 6))

# Boxplot grouped by MINORGROUP。箱线图
# 画 ATT8SCR 的箱线图（一条横向的箱线图，因为 x='ATT8SCR'，没有给 y 分组）。
    # data=england_school_2022_23：数据来源
    # color="#EDD971"：箱线图颜色
    # fliersize=0：不显示箱线图默认的离群点（避免和散点叠加太乱）
    # linewidth=1：线条粗细
sns.boxplot(
    x='ATT8SCR',
    data=england_school_2022_23,
    color="#EDD971",
    fliersize=0,
    linewidth=1
)

# Jittered points。叠加抖动散点
# 画每一所学校的 ATT8SCR 散点，并做抖动（jitter）让点不完全重叠。
    # hue='MINORGROUP'：用 MINORGROUP 给点上色（不同学校类型不同颜色）
    # dodge=False：不把不同 hue 的点分开摆放（都在同一条带上）
    # jitter=True：加随机抖动（避免点完全重合）
    # alpha=0.5：透明度 0.5
    # size=4：点大小
    # palette='turbo'：配色方案
sns.stripplot(
    x='ATT8SCR',
    data=england_school_2022_23,
    hue='MINORGROUP',
    dodge=False,
    jitter=True,
    alpha=0.5,
    size=4,
    palette='turbo'
)

# Customize the plot
plt.title("Attainment 8 - All Schools 2022/23 Academic Year")
plt.xlabel("Attainment 8 Score")
plt.ylabel("")
plt.legend(title="School Type", loc='lower center', bbox_to_anchor=(0.5, -0.25), ncol=2)
sns.despine()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Combine the datasets
# 把 camden_sub 和 leeds_sub 纵向拼接成一张表（行叠加）。默认不改索引。
camden_sub = england_filtered[england_filtered["LANAME"].eq("Camden")].copy()
leeds_sub  = england_filtered[england_filtered["LANAME"].eq("Leeds")].copy()

combined_data = pd.concat([camden_sub, leeds_sub])

# Create a function to make and customize the plot
def make_plot():
    g = sns.FacetGrid(combined_data, col="LANAME", height=5, aspect=1)    # 创建分面网格：按 LANAME 分成多个小图（这里会是两个小图：Camden 和 Leeds）。
    # 在每个分面里画散点：
        # x：贫困比例 PTFSM6CLA1A
        # y：Attainment 8 ATT8SCR
        # alpha：透明度
        # s：点大小
        # color="steelblue"：点颜色
    g.map_dataframe(
        sns.scatterplot,
        x="PTFSM6CLA1A",
        y="ATT8SCR",
        color="steelblue",
        alpha=0.7,
        s=60
    )
    g.set_axis_labels("% KS4 Pupils Disadvantaged (PTFSM6CLA1A)", "Attainment 8 Score (ATT8SCR)")
    g.fig.suptitle("Attainment 8 vs % Disadvantaged by Local Authority", fontsize=16)
    g.fig.tight_layout()
    g.fig.subplots_adjust(top=0.85)
    plt.show()

# Call the function once
make_plot()
```

```python
# 导入 GeoPandas（处理带几何列的地理数据）。matplotlib（画图）。colormap 模块（这段后面其实没用到 cm）。Normalize：把数值映射到 0–1 之间，方便套 colormap。contextily：用于加在线底图（OpenStreetMap 等）。
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
import contextily as ctx

plt.close("all")

# 把经纬度从 edubase_schools 拉回 camden_sub（列名按你 edubase 里实际存在的为准）
camden_sub2 = camden_sub.merge(
    edubase_schools[["urn", "easting", "northing"]],
    left_on="URN", right_on="urn", how="left"
)

sf_df_bng = gpd.GeoDataFrame(
    camden_sub2,
    geometry=gpd.points_from_xy(camden_sub2["easting"], camden_sub2["northing"]),
    crs="EPSG:27700"
)
# Reproject to Web Mercator for contextily。
# 把你的地理数据从当前坐标系（sf_df_bng 可能是英国国家网格 EPSG:27700）转换到 Web Mercator EPSG:3857。
# 这是因为 contextily 的底图要求 3857。
sf_df_mercator = sf_df_bng.to_crs(epsg=3857)

# Scale marker size。
# 取 TOTPUPS 的最小值。
min_val = sf_df_mercator['TOTPUPS'].min()
# 取 TOTPUPS 的最大值。
max_val = sf_df_mercator['TOTPUPS'].max()
sf_df_mercator['size_scale'] = 4 + (sf_df_mercator['TOTPUPS'] - min_val) * (12 - 4) / (max_val - min_val)

# Color mapping
norm = Normalize(vmin=sf_df_mercator['ATT8SCR'].min(), vmax=sf_df_mercator['ATT8SCR'].max())
cmap = plt.colormaps['magma']
sf_df_mercator['color'] = sf_df_mercator['ATT8SCR'].apply(lambda x: cmap(norm(x)))

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
sf_df_mercator.plot(ax=ax, color=sf_df_mercator['color'], markersize=sf_df_mercator['size_scale']**2, alpha=0.8)

```

```python
# 导入绘图和数值处理库。
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load your data into a DataFrame named england_filtered
# england_filtered = pd.read_csv("your_data.csv")

# Apply log10 transformation (replace 0s to avoid log(0))对 PTFSM6CLA1A 做 log10。因为 log10(0) 不存在，所以先把 0 替换成 NaN（缺失）。
england_filtered['log_PTFSM6CLA1A'] = np.log10(england_filtered['PTFSM6CLA1A'].replace(0, np.nan))
england_filtered['log_ATT8SCR'] = np.log10(england_filtered['ATT8SCR'].replace(0, np.nan))

# Drop rows with NaNs from log(0)同理，对 ATT8SCR 做 log10，0 先变 NaN。
england_filtered = england_filtered.dropna(subset=['log_PTFSM6CLA1A', 'log_ATT8SCR'])

# Set theme。设定 seaborn 的整体风格（深色网格背景）。
sns.set_theme(style="darkgrid")

# Create jointplot
# 画一个 联合图 jointplot：
    # 中间是散点图 + 回归线（kind="reg"）
    # 边上会自动给 x 和 y 的分布图（通常是直方/密度）
    # truncate=False：回归线画到数据范围边界
    # color="m"：主色调是品红（magenta）
    # height=7：图大小
g = sns.jointplot(
    x="log_PTFSM6CLA1A",
    y="log_ATT8SCR",
    data=england_filtered,
    kind="reg",
    truncate=False,
    color="m",
    height=7
)

# Label axes
g.set_axis_labels("log10(PTFSM6CLA1A)", "log10(ATT8SCR)")
```
